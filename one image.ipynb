```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_ge=ImageDataGenerator(rotation_range=45, width_shift_range=0.20,height_shift_range=0.20,shear_range=0.20,
    zoom_range=0.20,horizontal_flip=True,vertical_flip=True,fill_mode='nearest')
```

```python
img_ge
```

```python
c=0
```

```python
 for batch in img_ge.flow(img1, batch_size=25,save_to_dir="C:/Users/SANDEEP/Desktop/mb",save_prefix='sss',save_format='png'):
        c=c+1
        if c>=25:
            break
```
